## 🔁 Loops
Repeating code with `for` and `while` loops.

```python
# For loop example
print("Counting to 5:")
for i in range(1, 6):
    print(i)

# While loop example
count = 3
print("\nCountdown:")
while count > 0:
    print(count)
    count -= 1
print("Blastoff!")
```

## 📋 Lists
Storing collections of data.

```python
fruits = ["apple", "banana", "cherry"]

# Accessing items
print(f"First fruit: {fruits[0]}")

# Adding items
fruits.append("date")
print(f"All fruits: {fruits}")

# Looping through a list
for fruit in fruits:
    print(f"I like {fruit}")
```

## 🔧 Functions
Creating reusable blocks of code.

```python
def greet_user(name):
    """Greets the user by name."""
    return f"Hello, {name}!"

# Calling the function
message = greet_user("Student")
print(message)

def add_numbers(x, y):
    return x + y

result = add_numbers(5, 7)
print(f"5 + 7 = {result}")
```
