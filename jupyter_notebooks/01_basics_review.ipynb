## 🖨️ Output
Printing to the screen is done using the `print()` function.

```python
print("Hello, World!")
print("Welcome to Python review.")
```

## 🔤 Letters & Numbers
Working with strings (text) and integers/floats (numbers).

```python
name = "Gemini"
version = 1.5

# Using f-strings for formatting
print(f"I am {name} version {version}.")

# Basic math
a = 10
b = 3
print(f"{a} + {b} = {a + b}")
print(f"{a} divided by {b} is {a / b}")
```

## 🎲 Random
Generating random values using the `random` library.

```python
import random

# Random integer between 1 and 10
rand_num = random.randint(1, 10)
print(f"Random number: {rand_num}")

# Random float between 0.0 and 1.0
rand_float = random.random()
print(f"Random float: {rand_float}")
```

## ❓ Conditionals
Making decisions with `if`, `elif`, and `else`.

```python
temperature = 75

if temperature > 80:
    print("It's hot outside!")
elif temperature > 60:
    print("It's a nice day.")
else:
    print("It might be cold.")
```
