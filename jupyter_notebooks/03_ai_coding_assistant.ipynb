## 1. Setup: Creating Data
First, let's create a dummy CSV (Comma Separated Values) file to work with. You can ask AI to generate this setup code too!

```python
import csv

# Data to write
data = [
    ['Name', 'Grade'],
    ['Alice', '85'],
    ['Bob', '92'],
    ['Charlie', '78'],
    ['Diana', '95'],
    ['Evan', '60']
]

# Creating the file
with open('grades.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("grades.csv created successfully!")
```

## 2. The Prompt
To get AI to write the processing code, we need a clear prompt.

**Example Prompt:**
> *"Write a Python script using the csv library to read 'grades.csv'. Loop through the rows, print each student's name and grade, and calculate the class average. The CSV has a header row."*

Here is what that code might look like:

```python
import csv

print("--- Student Grades ---")

total_score = 0
count = 0

# NEW SKILL: Opening and reading a CSV file
with open('grades.csv', 'r') as file:
    reader = csv.reader(file)
    
    # Skip the header row
    header = next(reader)
    
    for row in reader:
        # Row is a list like ['Alice', '85']
        name = row[0]
        grade = int(row[1]) # Convert string to integer
        
        print(f"{name}: {grade}")
        
        total_score += grade
        count += 1

# Calculate average
if count > 0:
    average = total_score / count
    print(f"\nClass Average: {average:.2f}")
else:
    print("No data found.")
```

## 3. Refining with AI
We can ask the AI to improve the code. 

**Follow-up Prompt:**
> *"Refactor the previous code into a reusable function that takes the filename as an argument. Add error handling in case the file doesn't exist."*

```python
def process_grades(filename):
    try:
        with open(filename, 'r') as file:
            reader = csv.reader(file)
            next(reader) # Skip header
            
            grades = []
            for row in reader:
                grades.append(int(row[1]))
            
            if grades:
                avg = sum(grades) / len(grades)
                return avg
            else:
                return 0
                
    except FileNotFoundError:
        return "Error: File not found."

# Testing the function
result = process_grades('grades.csv')
print(f"Calculated Average: {result}")
```
